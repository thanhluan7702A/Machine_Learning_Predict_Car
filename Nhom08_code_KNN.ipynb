```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from math import sqrt
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in my_df.columns: 
  my_df[col] = le.fit_transform(my_df[col])

my_df
```

```python
ndf = my_df[['buying_price', 'maintenance_cost', 'doors', 'persons', 'lug_boot', 'safety', 'decision']].values
ndf
```

```python
#gia tri ngau nhien
point = np.array([1,1,3,2,0,0,3])
```

```python
def Euclidean_distance(row1, row2): 
    distance = 0
    for i in range(len(row1)-1):
        distance += (row1[i] - row2[i])**2  #(x1-x2)**2+(y1-y2)**2
    return sqrt(distance)
```

```python
def Get_Neighbors(train, test_row, k, neighbors):
    distance = list()
    data = []
    for i in train:
        dist = Euclidean_distance(test_row, i)
        distance.append(dist)
        data.append(i)
    distance = np.array(distance)
    data = np.array(data)
#tim index cua diem co khoang cach gan nhat
    index_dist = distance.argsort()
#tim du lieu cua index do
    data = data[index_dist]
    neighbors = neighbors[index_dist[:k]]
    print("Neighbors:", neighbors)
    return np.bincount(neighbors).argmax()

print('Label:',Get_Neighbors(ndf[:, :-1], point,8, ndf[:, -1]))
```
